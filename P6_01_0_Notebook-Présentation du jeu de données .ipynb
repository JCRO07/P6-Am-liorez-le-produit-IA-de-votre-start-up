```python
def load_json2(jpath):
    data_iterator = pd.read_json(jpath, chunksize=100000, lines=True)
    chunk_list = []
    for data_chunk in data_iterator:
        filtered_chunk = data_chunk.filter(items=['review_id', 'business_id', 'text', 'stars'])
        chunk_list.append(filtered_chunk)
        print('chunk appended')
        print()

    df = pd.concat(chunk_list)
    return df
```

```python
data2 = load_json2('yelp_academic_dataset_review.json')
```

```python
data2
```

# Sélection uniquement des avis insatisfactions

```python
data2 = data2[data2.stars < 3]
data2
```

# Jointure des avis d'insatisfactions en fonction des restaurants

```python
DataResto = pd.merge(data1, data2, how='inner', on=['business_id'])
DataResto
```
